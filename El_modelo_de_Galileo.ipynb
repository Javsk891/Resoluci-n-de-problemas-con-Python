#Modelo de Galileo

Una de las ecuaciones más utilizadas en el mundo de la Física mecánica es la ecuación de Galileo, quién cientos de años atras formulo una serie de plinomios que pudieron representar amtematicamente conceptos como la aceleración, la velocidad y la distancia. Si ignoramos el efecto de factores como el viento y la resistencia del aire, se pueden calcular las cantidades antes mencionadas para un objeto que inicia su trayectoria al nivel del suelo.

El modelo de Galileo dice que la posición horizontal de un objeto lanzado en movimiento parabolico en función del tiempo estaría dado por la expresión $x = v_1t$, donde $X$ sera la distancia horizontal, $V_1$ será la valocidad inicial con la que parte el objeto y $t$ será el tiempo transcurrido.

1) $x = v_1t$

Otra de las expresiones propuestas por Galileo, calcula la altura (posición vertical del objeto) en el tiempo t. $y = v_1t + \frac{at^2}{2}$ donde $v_2$ representa la velocidad inicial del objeto en el eje y (dirección de la gravedad) donde el valor de la gravedad quivale a -9.81 $\frac{m}{s^2}$ aproximadamente.

2) $y = v_1t + \frac{at^2}{2}$

Si sutituimos la primera ecuación en la segunda, podemos encontrar que la altura (posición veritcal) del objeto, respecto a la distancia recorrida (posición horizontal) estaría dada por la expresión $y = \frac{v_2}{v_1}x + \frac{ax^2}{2v_1^2}$.

3) $y = \frac{v_2}{v_1}x + \frac{ax^2}{2v_1^2}$

Podemos utilizar las ecuaciones de Galileo para modelar la trayectoria de un obketo bajo condiciones ideales en Python, como se puede apreciar en las siguientes líneas de código.

```python
def trayectoria(x): # Definimos la función de la trayectoria del objeto en función de x
  posición = 10*x - 5*(x**2) # Definimos la función que describirá la trayectoria de un objeto de estudio
  return(posición)
```

###Trayectoria del objeto

Podemos utilizar las herramientas de visualización de Python para observar como se puede ver aproximadamente la trayectoria del objeto. En este caso usaremos la librería de `matplotlib` para graficar, teniendo en cuenta que en este caso hipotetico se consideran condiciones ideales en donde se ignora la acción de factores naturales como la resistencia del aire y el viento.

La herramienta `matplotlib.pyplot` de la libreria `matplotlib` en una de los muchos instrumentos que nos ofrece este lenguaje de programación para eleaborar gráficos.

```python
import matplotlib.pyplot as plt # Importamos matplotlib.pyplot para realizar la gráfica
xs = [x / 100 for x in list(range(201))]  # Generamos una lista de valores en el eje x desde 0.0 hasta 2.0 en incrementos de 0.01.
ys = [trayectoria(x) for x in xs]  # Calculamos los valores correspondientes en el eje y usando la función 'trayectoria'.
plt.plot(xs, ys)  # Creamos una gráfica de línea con los valores de xs y ys.
plt.title('Trayectoria de un objeto lanzado')  # Establecemos el título de la gráfica.
plt.xlabel('Posición horizontal')  # Etiquetamos el eje x como "Posición horizontal".
plt.ylabel('Posición vertical')  # Etiquetamos el eje y como "Posición vertical".
plt.axhline(y=0)  # Añadimos una línea horizontal en y=0 para representar el nivel del suelo.
plt.show()  # Mostramos la gráfica generada en pantalla.
```

La gráfica 'Trayectoria de un objeto lanzado' muestra la trayectoria hipotetica que se espera que siga un objeto que es lanzado desde la superficie de la tierra bajo condiciones ideales, pero considerando claramente el efecto de la gravedad. Este movimiento se conoce en la física clásica como movimiento parabolico

#La estrategia de solución para x

Ahora que tenemos la acuación para la posición de x (el objeto lanzado), podemos resolver este problema para cualquier caso similar que sea de nuestro interes y que involucre las ecuaciones de Galileo. Por ejemplo podemos estar interesados en calcular la posición final de la trayectoria del objeto lanzado.

Este problema de interes podría ser solucionado perfectamente por medio de la ecuacuación de posición que se construyo antes.

###Solución para la trayectoria: $y = 10x - 5x^2$

Para este caso tenemos la trayectoria del objeto lanzado representada por la ecuación $y = 10x - 5x^2$, de antemano sabemos que el objeto es lanzado desde el nivel del suelo, por lo que la posición inicial debería ser sustituida por '0' dentro de dicha ecuacuón, de la siguiente manera.

$0 = 10x - 5x^2$

Entonces, podemos resolverla para x, usando la formula cuadratica o más conocida como la ecuación del estudiante.

$x = \frac{-b +- (b^2-4ac)^\frac{1}{2}}{2a}$

En este caso, encontramos que $x = 0$ y $x = 2$ son las soluciones para esta ecuación. La primera solución $x = 0$ es cuando el objeto inicia el movimietno y la segunda solución $x = 2$, cuando el objeto llega a la posición final.

Usando las ecuaciones de Galileo y la estrategia de solución para x, una maquina más sofiticada podría predecir la trayectoria de una pelota o de un misil e interceptarlo en medio de su recorrido.
